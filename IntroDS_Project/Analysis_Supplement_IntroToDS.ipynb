<h4 id='1_2'>Shapiro Wilk Test</h4>

 Shapiro Wilk Test is a test of null hypothesis that a sample is from population with a normal distribution. From the results below, the null hypothesis can be rejected, hence confirms the non-normality of sample. With 95% significance level, 95% of the samples would produce similar results.

```python
n = 500
alpha = 0.05
sample_rain = np.random.choice(turnstile_weather_rain['ENTRIESn_hourly'], size=n, replace=False)
sample_norain = np.random.choice(turnstile_weather_norain['ENTRIESn_hourly'], size=n, replace=False)
W_rain, p_rain = st.shapiro(sample_rain)
W_norain, p_norain = st.shapiro(sample_norain)
print "p_rain = {0}".format(p_rain)
print "p_norain = {0}".format(p_norain)
print "p_rain < {0}: {1}".format(alpha, (p_rain < alpha))
print "p_norain < {0}: {1}".format(alpha, (p_norain < alpha))
```

<h4 id='mann_u'> Mann Whitney U Test: </h4>


```python
with_rain_mean = np.mean(turnstile_weather_rain['ENTRIESn_hourly']) 
print "Mean of Rain group ridership: {0}".format(with_rain_mean)

without_rain_mean = np.mean(turnstile_weather_norain['ENTRIESn_hourly']) 
print "Mean of No-Rain group ridership: {0}".format(without_rain_mean)

U,p = st.mannwhitneyu(turnstile_weather_rain['ENTRIESn_hourly'], turnstile_weather_norain['ENTRIESn_hourly'] ) 
print "p-value [2 - sided]: {0} < 0.05 (p_critical) ".format(p*2)
    
```

<h3 id='average_p_various_samples'> Average value of p for various sample sizes: </h3>

To show that low p-values to reject null hypothesis of Mann-Whitney U test are obtained only on large sample sizes (>=3K) of each population. 

```python
n = 500
alpha = 0.05
sample_sizes = [50, 100, 500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 9585]
for n in sample_sizes:
    sample_rain = np.random.choice(turnstile_weather_rain['ENTRIESn_hourly'], size=n, replace=False)
    sample_norain = np.random.choice(turnstile_weather_norain['ENTRIESn_hourly'], size=n, replace=False)
    U,p = st.mannwhitneyu(sample_rain, sample_norain ) 
    print "p-value for n={1} [2 - sided]: {0} < 0.05 (p_critical) -- {2}".format(p*2,n ,(p  < alpha))

```

```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats as st
import statsmodels.api as sm
import scipy.optimize as op


import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("turnstile_weather_v2.csv")
df = df.reset_index(drop=True)
```

```python
df['ENTRIESn_hourly'].describe()
```

```python
df_row = df['ENTRIESn_hourly'].values

# Function to groupby column to Entriesn_hourly
def map_column_to_ENTRIESn_hourly(column):
    instances = column.values 
    
    entries_hourly = {} 
    for i in np.arange(len(instances)): 
        if instances[i] in entries_hourly:
            entries_hourly[instances[i]] += float(df_row[i])
        else:
            entries_hourly[instances[i]] = float(df_row[i])
            
    return entries_hourly

#Convert Dict to Dataframe
def get_df(dict_entries, column_name):
    df = pd.DataFrame(data=dict_entries.items(), columns=[column_name,'entries'])
    
    print "Sample 3"
    print df.head(3)
    print "\n\n"
    
    print "Describe : \n"
    print pd.DataFrame(df['entries']).describe()
    return df 

# Plot 
def plot_data(df, column_name, kindofplot, xaxis_labeled):
    
    
    if xaxis_labeled == True:
        df.plot(x=column_name, y='entries', title="{0} AND THEIR ENTRIES".format(column_name.upper()), kind=kindofplot, alpha=0.5, color='blue')
        plt.xlabel(column_name)
    else:
        df.plot(title="{0} AND THEIR ENTRIES".format(column_name.upper()), kind=kindofplot, alpha=0.5, color='blue')
        plt.xlabel("{0} row number".format(column_name))
        
    plt.ylabel('{0} entries'.format(column_name))
    plt.legend(['entries'])
    plt.show()
```

<h3 id='exploratory_analysis'>Exploratory Analysis:</h3>

```python
unit_entrieshourly = map_column_to_ENTRIESn_hourly(df['UNIT'])
unit_df = get_df(unit_entrieshourly, 'unit')
plot_data(unit_df, 'unit', 'line', False)
```

<h3 id='OLS'> Regression using OLS </h3>

```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats as st
import statsmodels.api as sm
import scipy.optimize as op

import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("turnstile_weather_v2.csv")
df = df.reset_index(drop=True)
station_DUMMY = pd.get_dummies(df['station'],prefix='station')
UNIT_DUMMY = pd.get_dummies(df['UNIT'],prefix='unit')

y = df['ENTRIESn_hourly']

x = df.copy()
df = df.join(station_DUMMY)
df = df.join(UNIT_DUMMY)

columns = ['hour','weekday','rain','fog','tempi']

x = df[columns]
x = x.join(station_DUMMY)
x = x.join(UNIT_DUMMY)

model = sm.OLS(y,x)
results = model.fit()

```

<h4 id='OLS_r2'> R-Squared: </h4>

```python
print results.rsquared
```

<h4 id='OLS_coeff'> Summary | Coefficients </h4>

```python
results.summary()
```

<h4 id='Residuals'> Residuals </h4>

```python
predictions = results.predict(x)
residuals = predictions - y
plt.scatter(predictions, residuals, alpha=0.7, color='blue', edgecolors='black')
plt.title('RESIDUAL PLOT')
plt.plot([np.min(predictions),np.max(predictions)], [0, 0], color='red')
plt.xlabel('predictions')
plt.ylabel('residuals')
plt.show()
```

##### Observations vs Predictions

```python
plt.hist(y, color='blue', alpha=0.5, label='observations')
plt.hist(predictions, color='green', alpha=0.7, bins=6, label='predictions')
plt.title('OBSERVATIONS vs PREDICTIONS')
plt.ylabel('frequency')
plt.legend()
plt.show()
```
